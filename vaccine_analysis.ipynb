```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('imunizacao_por_cidade.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
df.drop(columns='CRS', inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.value_counts('Município')
```

```python
new_df = df['Município'].str.split('-', n=1, expand=True)
df['novo_Município'] = new_df[1]
df.drop(columns='Município', inplace=True)
```

```python
new_df2 = df['% pop. pelo menos uma dose'].str.split('%', n=1, expand=True)
df['novo_% pop. pelo menos uma dose'] = new_df2[0]
df.drop(columns='% pop. pelo menos uma dose', inplace=True)
```

```python
new_df3 = df['% pop. esquema vacinal completo'].str.split('%', n=1, expand=True)
df['novo_% pop. esquema vacinal completo'] = new_df3[0]
df.drop(columns='% pop. esquema vacinal completo', inplace=True)

```

```python
df.head()
```

```python
df = df.rename(columns={'novo_Município':'Município', 'novo_% pop. pelo menos uma dose':'percento_pop_pelo_menos_uma_dose', 
'novo_% pop. esquema vacinal completo': 'percento_pop_esquema_vacinal_completo'})

df["percento_pop_esquema_vacinal_completo"] = df.percento_pop_esquema_vacinal_completo.astype(float)
df["percento_pop_pelo_menos_uma_dose"] = df.percento_pop_pelo_menos_uma_dose.astype(float)

df.info()
```

```python
df.to_csv('novo_imunizacao_por_cidade.csv')
```
